# Pandas Built-in Data Visualization

In this lecture we will learn about pandas built-in capabilities for data visualization! It's built-off of matplotlib, but it baked into pandas for easier usage!  

Let's take a look!

## Imports

```python
import numpy as np
import pandas as pd
%matplotlib inline
```

## The Data

There are some fake data csv files you can read in as dataframes:

```python
df1 =pd.read_csv('df1', index_col =0)
```

```python
df1.head()
```

```python
df2 =pd.read_csv('df2')
```

```python
df2.head()
```

```python
import seaborn as sns
df1['A'].hist(bins =30)
```

```python
df1['A'].plot(kind = 'hist', bins =30)
```

```python
df1['A'].plot.hist()
```

```python
df2.plot.area()
```

```python
df2.plot.area(alpha = 0.5)
```

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked = True)
```

```python
df2.plot.hist()
```

```python
df1.plot.line(x = df1.index, y ='B')
```

```python
df1.plot.scatter(x ='A', y= 'B', c = 'C', cmap = 'coolwarm', s =df1['D']*100)
```

```python
df2.plot.box()
```

```python
df = pd.DataFrame(np.random.randn(1000,2), columns=['a','b'])
```

```python
df.head()
```

```python
df.plot.hexbin(x ='a', y ='b',gridsize =25,cmap ='coolwarm')
```

```python
df2['a'].plot.kde()
```

```python
df2.plot.density()
```
